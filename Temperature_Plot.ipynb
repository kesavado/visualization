```python
import pandas as pd
import folium
import geopy # calling into the function
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim  # calling Nominatim which is OpenStreetMap data 
                                    # Nominatim uses OpenStreetMap data to find locations on Earth by name 
```

```python
df =pd.read_excel('city_temperatures.xls','Sheet1')
```

```python
df.head() #df.iloc[:,2]
```

```python
temp = pd.Series(df.iloc[:,2])
```

```python
for i in range (0,len(df.columns)):
    df.dropna(subset=[sorted(df)[i]],inplace=True)
```

```python
state = pd.Series(df.iloc[:,1])
```

```python
city = pd.Series(df.iloc[:,0])
```

```python
def co_ordinates(a): # defining the function for returning the co-ordinates
    geolocator = Nominatim(user_agent="MyApp")
    b = geolocator.geocode(a)
    return b.latitude,b.longitude
```

```python
co_ordinate = []  # collecting the co-ordinates
for i in range (0,len(city)):
    lat,long=co_ordinates(city[i])
    co_ordinate.append([lat,long]) # sending into the file
```

```python
lati=[]
longi=[]
for i in range(0,len(co_ordinate)):
    a=co_ordinate[i][0]
    lati.append(a)

for i in range(0,len(co_ordinate)):
    a=co_ordinate[i][1]
    longi.append(a)
```

```python
temperature_map=folium.Map(Location=[lat,long])
fg=folium.FeatureGroup(name='Temperatures')

for i in range(0,len(state)):
    fg.add_child(folium.Marker(location=[co_ordinate[i][0],co_ordinate[i][1]],popup=city[i]+","+str(temp[i]),icon=folium.Icon(color='green')))
temperature_map.add_child(fg)
```

```python
fg.add_child(HeatMap(list(zip(lati,longi,temp)),min_opacity=0.2,max_val=temp.max(),
                     radius=20,blur=30))
temperature_map.add_child(fg)

#temperature_map.save('class_activity_2.html')  # for saving the the output as html file.
```
